# recall 데이터에 맞게 컬럼 네임 변경

```python
df_com.rename(columns={'COMPDESC':'COMPNAME'},inplace=True)
df_inv.rename(columns={'COMP_NAME':'COMPNAME','MAKE':'MAKETXT','YEAR':'YEARTXT','MODEL':'MODELTXT'},inplace=True)
```

# 조사데이터 결측지 제거 후 기준 컬럼만 남기기 

```python
di = df_inv[['MAKETXT','MODELTXT','YEARTXT','COMPNAME']]

for i in range(len(di)):
    if len(di.loc[i]['MAKETXT']) == 0:
        di.loc[i]['MAKETXT'] = np.nan
```

```python
di = di.dropna(subset='MAKETXT')
di = di.drop_duplicates()
```

```python
di
```

```python
dr = df_recall[['MAKETXT','MODELTXT','YEARTXT','COMPNAME']]
dr = dr.drop_duplicates()
dr
```

# 기준 컬럼으로 inner join

```python
df_merge_inv = pd.merge(df_com,di,on=['MAKETXT','MODELTXT','YEARTXT','COMPNAME'],how='inner')
df_merge_inv
```

```python
inv_to_rcl = pd.merge(df_merge_inv,dr,on=['MAKETXT','MODELTXT','YEARTXT','COMPNAME'],how='inner')
```

```python
inv_to_rcl
```

```python
df_com['label'] = 0
```

```python
cmpl_list = inv_to_rcl['CMPLID'].values.tolist()
cmpl_list.sort()
```

```python
cmpl_list = [int(x) for x in cmpl_list]
```

```python
for i in cmpl_list:
    df_com.loc[[i-1],['label']] = 1
```

```python
df = df_com[['MAKETXT','MODELTXT','YEARTXT','COMPNAME','INJURED','DEATHS',
             'DATEA','LDATE','MILES','OCCURENCES','CDESCR','PURCH_DT',
            'ORIG_EQUIP_YN','MANUF_DT','PROD_TYPE','label']]




```
